```python
# a script that evaluates each line in a list of potential crossword puzzle entries
# to locate entries of at least 7 characters whose first three characters match their fourth through sixth characters
# as in the name of the infamous "Star Wars Episode 1" character Jar Jar Binks

# crossword entry lists are generally text files in which each line contains
# the entry, followed by a semicolon, then a score the construction program algorithm uses to
# sort potential entries, e.g. "entry;50"
# this function strips the word scores from the entries
# and returns entries of the desired length

def shorten(f, entry_length):
  shortened_output = []
  with open(f) as list:
    for l in list:
      x = l[0:l.find(";")]
      if len(x) == int(entry_length):
        shortened_output.append(x)
  return(shortened_output)

# this function eliminates 7-character entries
# that are simply plurals of 6-character entries
# and entries of 9 or more characters that continue to repeat
# the first 3-character substring, e.g. "saysaysay"
# or "idoidoido"
def mere_plural_or_dupe(x):
  if len(x) == 7 and x[-1] =="s":
    return True
  elif len(x) >= 9 and x[6:9] == x[0:3]:
    return True
  else:
    return False

def main():

  # insert link to word list here
  wdlist = "#"

  # empty list to which potential entries will be appended
  out = []

  # crossword entries are always at least 3 characters
  # and the standard crossword grid size
  # is a maximum of 16 by 16 characters
  for x in range(7, 16):
    ents = shorten(wdlist, x)

    for i in range(len(ents)):
      y = ents[i]
      if y not in out:
        if y[3:6] == y[0:3] and mere_plural_or_dupe(y) == False:
          out.append(y)

  for z in range(len(out)):
    print(out[z])

main()
```
